```python
df1 = pd.DataFrame(data,columns=['Season','Player','Liga_Goals','Liga_Asts','Liga_Aps','Liga_Mins','CL_Goals','CL_Asts','CL_Aps','CL_Mins'])
data_of_r = df1[df1['Player']=='Ronaldo']
s = pd.Series([0,1,2,3,4,5,6,7,8])
data_of_r = data_of_r.set_index([s])
#data_of_r.drop(data_of_r['index'],inplace=True)
data_of_r
```

```python
df2 = pd.DataFrame(data,columns=['Season','Player','Liga_Goals','Liga_Asts','Liga_Aps','Liga_Mins','CL_Goals','CL_Asts','CL_Aps','CL_Mins'])
data_of_m = df2[df2['Player']=='Messi']
s = pd.Series([0,1,2,3,4,5,6,7,8])
data_of_m = data_of_m.set_index([s])
data_of_m
```

```python
messi_stats_LaLiga = pd.DataFrame(data_of_m,columns=['Goals_per_Game','Assists_per_Game','Minutes_per_Goal','Total_Contribution_bym','NumberOfGames'])
messi_stats_LaLiga['Goals_per_Game'] = data_of_m['Liga_Goals']/data_of_m['Liga_Aps']
messi_stats_LaLiga['Assists_per_Game'] = data_of_m['Liga_Asts']/data_of_m['Liga_Aps']
messi_stats_LaLiga['Minutes_per_Goal'] = data_of_m['Liga_Mins']/data_of_m['Liga_Goals']
messi_stats_LaLiga['Total_Contribution_bym'] = data_of_m['Liga_Goals']+data_of_m['Liga_Asts']
messi_stats_LaLiga['NumberOfGames'] = data_of_m['Liga_Aps']
messi_stats_LaLiga
```

```python
ronaldo_stats_LaLiga = pd.DataFrame(data_of_r,columns=['Goals_per_Game','Assists_per_Game','Minutes_per_Goal','Total_Contribution_byr','NumberOfGames'])
ronaldo_stats_LaLiga['Goals_per_Game'] = data_of_r['Liga_Goals']/data_of_r['Liga_Aps']
ronaldo_stats_LaLiga['Assists_per_Game'] = data_of_r['Liga_Asts']/data_of_r['Liga_Aps']
ronaldo_stats_LaLiga['Minutes_per_Goal'] = data_of_r['Liga_Mins']/data_of_r['Liga_Goals']
ronaldo_stats_LaLiga['Total_Contribution_byr'] = data_of_r['Liga_Goals']+data_of_r['Liga_Asts']
ronaldo_stats_LaLiga['NumberOfGames'] = data_of_r['Liga_Aps']
ronaldo_stats_LaLiga
```

```python
messi_hist = messi_stats_LaLiga.hist(column='Total_Contribution_bym')
messi_hist
```

```python
ronaldo_hist = ronaldo_stats_LaLiga.hist(column='Total_Contribution_byr')
ronaldo_hist
```

```python
#let's compare them year by year
comparison_yearwise = pd.DataFrame(ronaldo_stats_LaLiga,columns=['Year','Total_Contribution_bym','Total_Contribution_byr','Winner'])
comparison_yearwise['Year'] = data_of_m['Season']
comparison_yearwise['Total_Contribution_bym']=data_of_m['Liga_Goals']+data_of_m['Liga_Asts']
comparison_yearwise['Total_Contribution_byr']=ronaldo_stats_LaLiga['Total_Contribution_byr']
comparison_yearwise['Winner']= ronaldo_stats_LaLiga['Total_Contribution_byr'] > messi_stats_LaLiga['Total_Contribution_bym'] 

comparison_yearwise

```
